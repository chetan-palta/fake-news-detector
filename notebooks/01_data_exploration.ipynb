```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Datasets
fake = pd.read_csv('../data/dataset/Fake.csv')
real = pd.read_csv('../data/dataset/True.csv')
```

```python
# Add Labels to Both
fake['label'] = 0  # fake
real['label'] = 1  # real
```

```python
# Merge Them
df = pd.concat([fake, real]).reset_index(drop=True)
```

```python
# Basic EDA Checks
print(df.shape)
print(df['label'].value_counts())
df.head()
df.info()
```

```python
# Plot Class Balance
sns.countplot(x='label', data=df)
plt.xticks([0,1], ['Fake', 'Real'])
plt.title("Class Distribution")
plt.show()
```
